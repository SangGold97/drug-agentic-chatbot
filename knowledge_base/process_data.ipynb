```python
import pandas as pd
import os
import numpy as np
```

```python
drug = pd.read_csv('drug_info.csv')
drug.drop(columns=['Diseases', 'Products'], axis=1, inplace=True)
drug
```

```python
additional_info = pd.read_csv('drug_additional_info.csv')
additional_info.rename(columns={'related diseases': 'related_diseases',
                                'product names': 'product_names'}, inplace=True)
additional_info
```

```python
drug_combined = pd.concat([drug, additional_info[['related_diseases', 'product_names']]], axis=1)
drug_combined
```

```python
# Create a new column 'related_gene' initialized with empty strings
drug_combined['related_gene'] = ""

# Loop through each file in the reference directory
reference_dir = 'reference'
for filename in os.listdir(reference_dir):
    if filename.endswith('.xlsx'):
        # Get the file name (remove .xlsx extension), split by space, take the first element, convert to lowercase
        file_name = filename.replace('.xlsx', '').split()[0].lower()

        # Read the first sheet of the xlsx file
        file_path = os.path.join(reference_dir, filename)
        df = pd.read_excel(file_path, sheet_name=0)

        # Get information from the 'Gene_list' column, first row (index 0)
        if 'Gene_list' in df.columns and len(df) > 0:
            gene_info = str(df['Gene_list'].iloc[0])
            # Split by '@' and take the first element
            gene = gene_info.split('@')[0]

            # Find rows where drug name starts with file_name and update related_gene
            mask = drug_combined['name'].str.lower().str.startswith(file_name)
            drug_combined.loc[mask, 'related_gene'] = gene

print(f"Processed {len(os.listdir(reference_dir))} files in the reference directory")
drug_combined.head()
```

```python
drug_combined.to_csv('knowledge_base.csv', index=False)
```
